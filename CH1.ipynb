```python
import random as rnd

kilometers = list(range(1,101))
values = dict()

for i in range(50):
    choice = rnd.choice(kilometers)
    values[choice] = choice  * 0.621371
    
    #0.621371
```

```python
weight = 0.5
alpha = 0.0001

iteration = 1
for i in range(1):
    for kilometer, correct in values.items():
        mile = weight * kilometer
        error = (mile - correct)**2
        delta = mile - correct
        weight_delta = kilometer * delta
        weight -= weight_delta * alpha




        print(f"Iter: {iteration:4} | Predicition {mile:12.4} | Error {error:10.4} | Delta {delta:10.4}")
        iteration += 1
```

```python
worms = []
bugs = []
for i in range(500):
    x = rnd.uniform(1, 2)
    y = rnd.uniform(4,7)
    worms.append((x,y))
    x = rnd.uniform(2, 7)
    y = rnd.uniform(1, 4)
    bugs.append((x,y))
insects = worms + bugs

rnd.shuffle(insects)
```

```python
l = 0.0002
a = 1

for i in range(12):
    for item in insects:
        x = item[0]
        y = item[1]
        t = a * x 
        e = y - t
        deltaA = e/x
        a += deltaA * l

        print(f"X:{x:5.3} Y:{y:5.3} | T: {t:10.3} | E: {e:12.4} | A: {a:12.4} | Delta A: {deltaA:12.4}")
    print("*"*100)
   
```

```python
import matplotlib.pyplot as plt
x,y = [],[]
for i,j in insects:
    x.append(i)
    y.append(j)
    
ly = [lx * a for lx in x]

plt.scatter(x,y, c = 'g', s=4)
plt.plot(x, ly, c = "r")

plt.show()
print(a)
```
