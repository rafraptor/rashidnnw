```python
import random as rnd

kilometers = list(range(1,101))
values = dict()

for i in range(50):
    choice = rnd.choice(kilometers)
    values[choice] = choice  * 0.621371
    
    #0.621371
```

```python
weight = 0.5
alpha = 0.0001

iteration = 1
for i in range(1):
    for kilometer, correct in values.items():
        mile = weight * kilometer
        error = (mile - correct)**2
        delta = mile - correct
        weight_delta = kilometer * delta
        weight -= weight_delta * alpha




        print(f"Iter: {iteration:4} | Predicition {mile:12.4} | Error {error:10.4} | Delta {delta:10.4}")
        iteration += 1
```

```python
worms = []
bugs = []
for i in range(300):
    x = rnd.uniform(1, 4)
    y = rnd.uniform(4,7)
    worms.append((x,y))
    x = rnd.uniform(4, 7)
    y = rnd.uniform(1, 4)
    bugs.append((x,y))
insects = worms + bugs

rnd.shuffle(insects)
```

```python
l = 0.01
a = 0.5

for item in insects:
    x = item[0]
    y = item[1]
    t = a * x
    e = t - y
    deltaA = e/x
    a += deltaA
    
    print(f"X:{x:4.2} Y:{y:4.2} | T: {t:10.2} | E: {e:12.2} | A: {a:12.2} | Delta A: {deltaA:12.4}")
```

```python
import matplotlib.pyplot as plt
x,y = [],[]
for i,j in insects:
    x.append(i)
    y.append(j)
    
lx = [0, 6.8]
ly = [0, 6.8*1]

plt.scatter(x,y, c = 'g', s=2)
plt.plot(lx, ly, c = "r")

plt.show()
```
